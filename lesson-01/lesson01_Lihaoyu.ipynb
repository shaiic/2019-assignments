```python
a = [1,2,24,6,101,1234567]           #创建数组1，可更改
b = [2,4,6,56,10,10086]          #创建数组2，可更改
c = a + b                 #合并数组
print(c)                  
list.sort(c)              #做sort
print(c)
```

```python
a = list(range(101))   #创建0-100的数组
print(a)                    
import random          #导入random函数
i = random.randint(1,100)  #取随机数
print(i)
a[0] = a[i]            #替换0
print(a)
random.shuffle(a)      #随机打乱
print(a)

list.sort(a)           #做sort
while len(a) > 1:      #循环条件，list a的长度大于1
    if a[0] == a[1]:   #判断第一、第二个元素是否相等
        print(a[0])   
        break          #如果相等，打断循环
    else:
        del a[0]       #如果不相等，去掉第一个元素，重新循环
```

```python
a = list(range(101))   #创建0-100的数组
print(a)                    
import random          #导入random函数
i = random.randint(1,100)  #取随机数
print(i)
a[0] = a[i]            #替换0
random.shuffle(a)      #随机打乱
print(a)

S = sum (a)            #求和
b = S - 5050           #减去0到100的和
print(b)               #得到结果
```
