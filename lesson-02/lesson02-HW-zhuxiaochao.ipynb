````python
import numpy as np
np.random.seed(42)
height_10k = np.random.normal(1.560, 0.0507, 10000)   # 10000èº«é«˜æ•°æ®   
#hzFemale_height = height_10k[:,0]  #10000åä¸­å¥³æ€§æ•°æ®èº«é«˜å€¼
hzFemale_height_var = np.var(height_10k)    # èº«é«˜æ–¹å·®
hzFemale_height_mean = np.mean(height_10k)  # èº«é«˜å‡å€¼
hzFemale_height_median = np.median(height_10k)    # èº«é«˜ä¸­ä½æ•°
print('åä¸­å¥³æ€§èº«é«˜æ–¹å·®'+str(hzFemale_height_var*(1000**2)))
print('åä¸­å¥³æ€§èº«é«˜å‡å€¼'+str(hzFemale_height_mean*1000))
print('åä¸­å¥³æ€§èº«é«˜ä¸­ä½æ•°'+str(hzFemale_height_median*1000))
#ç­”æ¡ˆæ˜¯C
````

### **Q6.å¹´é¾„å°äº30å²çš„ä¹˜å®¢æ˜¯å¦æ¯”å¹´é¾„å¤§äº60å²çš„ä¹˜å®¢æœ‰æ›´å¤§çš„å­˜æ´»æ¦‚ç‡ï¼Ÿ å­˜æ´»æ¦‚ç‡å„ä¸ºå¤šå°‘ï¼Ÿ**
