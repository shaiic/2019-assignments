#### é—®ï¼š ä½ ç”Ÿæˆçš„åˆ†å¸ƒçš„å®é™…äº§ç”Ÿçš„æ–¹å·®ã€å‡å€¼ã€ä¸­ä½æ•°å„æ˜¯å¤šå°‘(ä¿ç•™ä¸¤ä½å°æ•°)ï¼Ÿ

````python
import numpy as np
np.random.seed(42)
# your code
height_10k = np.random.normal(1560, 50.7, 10000)   # 10000èº«é«˜æ•°æ®
weight_10k = np.random.normal(50, 6.8, 10000)         # 10000ä½“é‡æ•°æ®
shmale_10k = np.column_stack((height_10k, weight_10k)) # 10000åä¸­æ•°æ®
shmale_10k                                           # éšæœºäº§ç”Ÿä¸€äº›ç˜¦å­å’Œèƒ–å­

shmale_weight = shmale_10k[:,1]
shmale_height = shmale_10k[:,0]
shmale_height_mean = np.mean(shmale_height)  # èº«é«˜å‡å€¼
shmale_height_std = np.std(shmale_height)    # èº«é«˜æ ‡å‡†å·®
shmale_height_var = np.var(shmale_height)    # èº«é«˜æ–¹å·®
shmale_height_median = np.median(shmale_height)    # èº«é«˜ä¸­ä½æ•°
shmale_weight_mean = np.mean(shmale_weight)  # ä½“é‡å‡å€¼
shmale_weight_std = np.std(shmale_weight)    # ä½“é‡æ ‡å‡†å·®
shmale_weight_var = np.var(shmale_weight)    # ä½“é‡æ–¹å·®
shmale_weight_median = np.median(shmale_weight)    # ä½“é‡ä¸­ä½æ•°
from tabulate import tabulate                # æ ¼å¼åŒ–æˆè¡¨æ ¼æ ·å¼
print(tabulate([['èº«é«˜ï¼ˆç±³ï¼‰', shmale_height_mean, shmale_height_std,shmale_height_var,shmale_height_median], 
                ['ä½“é‡ï¼ˆå…¬æ–¤ï¼‰', shmale_weight_mean, shmale_weight_std,shmale_weight_var,shmale_weight_median]], 
               headers=['åä¸­å¥³','å‡å€¼', 'æ ‡å‡†å·®','æ–¹å·®','ä¸­ä½æ•°']))
````
